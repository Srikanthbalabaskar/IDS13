### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
print(users.head())
```

### Step 4. Discover what is the mean age per occupation

```python
mean_age = pd.Series((users.groupby('occupation')['age'].mean()), name="Mean age")
pd.concat([mean_age], axis=1)
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users1 = users.groupby(['occupation'])['gender'].count()
total_people = pd.Series(users1, name="Total Resources")
male_users = users.query('gender == "M"').groupby('occupation').count()['gender']
male_users = pd.Series(male_users, name="Male_users")
male_user_ratio = (male_users / total_people) * 100
male_user_ratio = pd.Series((male_user_ratio.sort_values(ascending=False)), name="Male/Total Ratio")
pd.concat([male_users, total_people, male_user_ratio], axis=1)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
min_series = users.groupby('occupation')['age'].min()
min_series = pd.Series(min_series, name='Min age')
max_series = users.groupby('occupation')['age'].max()
max_series = pd.Series(max_series, name = 'Max age')
pd.concat([min_series, max_series], axis=1)
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
gender_occupation = users.groupby(['occupation', 'gender'])['age'].mean()
gender_occ_series = pd.Series(gender_occupation, name = "Mean age")
pd.concat([gender_occ_series], axis=1)
```

### Step 8.  For each occupation present the percentage of women and men

```python
male_users = users.query('gender == "M"').groupby(['occupation']).count()['gender']
female_users = users.query('gender == "F"').groupby(['occupation']).count()['gender']
total_users = users.groupby('occupation').count()['gender']
male_percentage = (male_users / total_users) * 100
male_series = pd.Series(male_percentage, name="Male %")
female_percentage = (female_users / total_users) * 100
female_series = pd.Series(female_percentage, name="Female %")
pd.concat([male_series, female_series], axis=1)
```
