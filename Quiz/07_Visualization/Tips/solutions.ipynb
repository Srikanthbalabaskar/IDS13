### Step 5. Plot the total_bill column histogram

```python
bill_series = tips['total_bill']
```

```python
bill_series = pd.Series(bill_series, name='total bill')
bill_df = pd.DataFrame(bill_series)
```

```python
sns.distplot(bill_df)
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
bill_tip = tips[['total_bill', 'tip']]
sns.scatterplot(x='total_bill', y='tip', data=bill_tip)
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
flatui = ["#9b59b6", "#2ecc71", '#EA6A47','#1C4E80', '#0091D5','#B3C100', '#4CB5F5', '#1F3F49', '#D32D41', '#6AB187', '#23282D','#A5D8DD','#7E909A']
sns.palplot(sns.color_palette(flatui))
```

```python
bill_tip_size = tips[['total_bill', 'tip', 'size']]
sns.set_style('darkgrid')
sns.scatterplot(x='total_bill', y='tip', data=bill_tip_size, size='size', hue='size', palette=sns.color_palette(flatui, n_colors=6))
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
bill_day = tips[['total_bill', 'day']]
sns.barplot(x='day', y='total_bill', data=bill_day)
plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
day_tip = tips[['tip', 'day','sex']]
sns.scatterplot(x='tip', y='day', data=day_tip, hue='sex')
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
bill_day_time = tips[['total_bill', 'day', 'time']]
bill_day_time.groupby(['day','time']).sum()
```

```python
sns.boxplot(x = 'day', y='total_bill', data=bill_day_time, hue='time')
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
tip_day_df = tips[['tip','time']]
tip_day_dinner_df = tip_day_df[tip_day_df['time'] == 'Dinner']['tip']
tip_day_lunch_df = tip_day_df[tip_day_df['time'] == 'Lunch']['tip']

fig, ax = plt.subplots(1,2)
sns.distplot(tip_day_dinner_df, ax=ax[0], axlabel="Dinner-tip")
sns.distplot(tip_day_lunch_df, ax=ax[1], axlabel="Lunch-tip")
plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
tips.head()
```

```python
bill_gender_tip = tips[['total_bill','tip','sex','smoker']]
bill_male_tip = bill_gender_tip[bill_gender_tip['sex'] == 'Male']
bill_female_tip = bill_gender_tip[bill_gender_tip['sex'] == 'Female']

fig, ax = plt.subplots(1,2)
sns.scatterplot(x='total_bill', y='tip', data=bill_male_tip, ax=ax[0], hue='smoker' )
sns.scatterplot(x='total_bill', y='tip', data=bill_female_tip, ax=ax[1], hue='smoker')
plt.show()
```
