### Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.

```python
data = pd.read_fwf('wind.data')
data.head()
```

```python
# replace the first row as the columns
data.columns = data.iloc[0]
data = data[1:]
data.head()
```

```python
# changing the year month day column to a datetime column
# to do that we need to use the to_datetime function
# for to_datetime to work, the columns have to have names as 'year', 'month', 'day'

data = data.rename(columns = {
    'Yr': 'year',
    'Mo': 'month',
    'Dy': 'day'
})
```

```python
data['year'] = '19' + data['year']
```

```python
data['DateTime'] = pd.to_datetime(data[['year', 'month','day']])
```

```python
del data['year']
del data['month']
del data['day']
```

```python
data = data.set_index('DateTime')
data.head()
```

### Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day.

#### A different set of numbers for each day.

```python
min_series = data1.min(axis=1)
max_series = data1.max(axis=1)
mean_series = data1.mean(axis=1)
std_series = data1.std(axis=1)

# put a name for each series

min_series = pd.Series(min_series, name = 'Min')
max_series = pd.Series(max_series, name='Max')
mean_series = pd.Series(mean_series, name = 'mean')
std_series = pd.Series(std_series, name = 'std_dev')

day_stats = pd.concat([min_series, max_series, mean_series, std_series], axis=1)
day_stats
```
