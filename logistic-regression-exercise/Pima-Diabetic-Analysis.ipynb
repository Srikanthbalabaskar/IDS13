```python
# function to fill the respective glucose level based on age
def glucose_by_age(cols):
    Glucose = cols[0]
    Age = cols[1]
    if Glucose == 0:
        if Age == 21 or Age == 22:
            return 108
        if Age == 37:
            return 130
        if Age == 41:
            return 129
    else:
        return Glucose
```

```python
pima_df['Glucose'] = pima_df[['Glucose', 'Age']].apply(glucose_by_age, axis=1)
```

```python
# imputing the insulin values with the mean glucose levels 
# in a certain range seems to be a good idea

# first we need to split the glucose levels into different bins
pd.cut(pima_df['Glucose'], bins=6).value_counts()
```

```python
# we can find the mean in each of these ranges and then impute the values accordingly

bin_95_121 = pima_df[(pima_df.Glucose > 95) & (pima_df.Glucose <= 121)]['Glucose'].mean()
bin_121_147 = pima_df[(pima_df.Glucose > 121) & (pima_df.Glucose <= 147)]['Glucose'].mean()
bin_69_95 = pima_df[(pima_df.Glucose > 69) & (pima_df.Glucose <= 95)]['Glucose'].mean()
bin_147_173 = pima_df[(pima_df.Glucose > 147) & (pima_df.Glucose <= 173)]['Glucose'].mean()
bin_173_199 = pima_df[(pima_df.Glucose > 173) & (pima_df.Glucose <= 199)]['Glucose'].mean()
bin_43_69 = pima_df[(pima_df.Glucose > 43) & (pima_df.Glucose <= 69)]['Glucose'].mean()
bin_173_199
```

```python
# write a function to impute insulin levels

def impute_insulin(cols):
    Glucose = cols[0]
    Insulin = cols[1]
    if Insulin == 0:
        if Glucose > 43 and Glucose <= 69:
            return 61
        if Glucose > 69 and Glucose <= 95:
            return 86
        if Glucose > 95 and Glucose <= 121:
            return 108
        if Glucose > 121 and Glucose <= 147:
            return 133
        if Glucose > 147 and Glucose <= 173:
            return 160
        if Glucose > 173 and Glucose <= 199:
            return 186
    else:
        return Insulin
    
```

```python
pima_df['Insulin'] = pima_df[['Glucose', 'Insulin']].apply(impute_insulin, axis=1)
```

```python
# Now that all the values have been imputed properly, we can proceed to train the model.
```

```python
# Lets split the dataset into train and validation datasets

train_test = pima_df[0:699]
validation = pima_df[700:]
```

```python
train_test.shape
```

```python
validation.shape
```

```python
X = train_test.drop('Outcome', axis=1)
y = train_test['Outcome']
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(X_train, y_train)
```

```python
log.intercept_
```

```python
log.coef_
```

```python
predict = log.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix,roc_auc_score
```

```python
print(accuracy_score(y_test,predict))
```

```python
print(confusion_matrix(y_test,predict))
```

```python
print(roc_auc_score(y_test,predict))
```

```python
# try and change the decision boundary
pred_prob = log.predict_proba(X_test)
```

```python
pred_new = np.where(pred_prob[:,0]>=0.62,0,1)
print(accuracy_score(y_test,pred_new))
print(confusion_matrix(y_test,pred_new))
print(roc_auc_score(y_test,pred_new))
```

```python
# lets check with the validation dataset how our model is performing

validation_x = validation.drop('Outcome', axis=1)
validation_y = validation['Outcome']
```

```python
pred_validation = log.predict(validation_x)
```
